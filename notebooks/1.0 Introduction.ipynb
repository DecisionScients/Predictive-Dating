# Introduction
For millions of years, humans have been selecting partners based upon a wealth of information gleaned in face-to-face interactions — not just appearance, but characteristics such as tone of voice, body language, scent, as well as social status. In contemporary Western societies, this selection decision usually follows a long period during which people engage in more informal, primarily polygamous relationships, i.e, dating. But how does the selection process manifest in a speed dating environment?  

This study examines the dating preferences of subjects who have participated in a number (between 9 and 21), of four minute "speed dates". After which, each subject has the opportunity to accept or reject each partner. The purpose of this study is to illucidate dating behavior and selection preferences along several axis:
1. Gender - Gender differences have been the subject of many studies over the years. The goal is to illucidate new patterns of behavior that are perhaps unique to this context.
2. Values - Subjects rated each other and themselves along six characteristics: attractiveness, sincerity, intelligence, fun, ambition, and shared interests. To what degree do these attributes correlate?
3. Self-Perception / Awareness - How does self-perception shape dyadic interactions? What about agreement/disagreement in assessments? 
4. Race - What are the racial dynamics exhibited in the speed dating environment?

Once the exploratory data analysis is complete, a prediction model will be developed to predict both individual selection as well as matches between subjects. 

## Data
This data was gathered from participants in experimental speed dating events from 2002-2004. During the events, the attendees would have a four-minute "first date" with every other participant of the opposite sex. At the end of their four minutes, participants were asked if they would like to see their date again. They were also asked to rate their date on six attributes: Attractiveness, Sincerity, Intelligence, Fun, Ambition, and Shared Interests. The dataset also includes questionnaire data gathered from participants at different points in the process. These fields include: demographics, dating habits, self-perception across key attributes, beliefs on what others find valuable in a mate, and lifestyle information.
